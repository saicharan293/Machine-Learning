## Black Friday EDA and Feature Engineering
### Cleaning and Preparing the data for Model training

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
#importing dataset
df_train=pd.read_csv('train.csv')
df_train.head()
df_train.shape
```

```python
df_test=pd.read_csv('test.csv')
df_test.shape
```

```python
## Merge train and test data
df=df_train._append(df_test)
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
## user id isn't needed in above df, so it need to be removed
df.drop(['User_ID'],axis=1,inplace=True)
df.head()
```

## Fix the categorical Feature to Numerical

```python
pd.get_dummies(df['Gender']).astype(int)
```

```python
# Handling Categorical feature : Gender
df['Gender']=df['Gender'].map({'F':0,'M':1})
df.head()
```

```python
# Handling categorical feature : AGE
df['Age'].unique()
```

```python
df['Age']=df['Age'].map({'0-17':1,'18-25':2,'26-35':3,'36-45':4,'46-50':5,'51-55':6,'55+':7})
df.head()
```

```python
# Handling Categorical feature : city category
df_city=pd.get_dummies(df['City_Category'],drop_first=True).astype(int)
df_city.head()
```

```python
df=pd.concat([df,df_city],axis=1)
df.head()
```

```python
#drop city category
df.drop('City_Category',axis=1,inplace=True)
df.head()
```

## Check the missing values

```python
df.isnull().sum()
```

### Replace missing values for product category

```python
df['Product_Category_2'].unique()
```

```python
mode_pr=df['Product_Category_2'].mode()[0]
mode_pr
```

```python
df['Product_Category_2']==df['Product_Category_2'].fillna(df['Product_Category_2'].mode()[0],inplace=True)
```

```python
df['Product_Category_3'].isnull().sum()
```

```python
## Product_Category_3 : replace missing values
df['Product_Category_3'].fillna(df['Product_Category_3'].mode()[0],inplace=True)
```

```python
df.head()
```

```python
# Stay_In_Current_City_Years
df['Stay_In_Current_City_Years'].unique()
```

```python
df['Stay_In_Current_City_Years']=df['Stay_In_Current_City_Years'].str.replace('+','')
```

```python
df.info()
```

```python
## convert Stay_In_Current_City_Years (object) into integer
df['Stay_In_Current_City_Years']=df['Stay_In_Current_City_Years'].astype(int)
```

```python
df[['B','C']]=df[['B','C']].astype(int)
```
