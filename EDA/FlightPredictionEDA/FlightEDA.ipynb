### Feature Engineering

```python
final_df['Date']=final_df['Date_of_Journey'].str.split('/').str[0]
final_df['Month']=final_df['Date_of_Journey'].str.split('/').str[1]
final_df['Year']=final_df['Date_of_Journey'].str.split('/').str[2]
```

```python
final_df.head(2)
```

```python
final_df.info()
```

```python
final_df['Date']=final_df['Date'].astype(int)
final_df['Month']=final_df['Month'].astype(int)
final_df['Year']=final_df['Year'].astype(int)
```

```python
final_df.info()
```

```python
final_df.drop('Date_of_Journey',axis=1,inplace=True)
final_df.head()
```

### arrival time

```python
# final_df['Arrival_Time'].str.split(' ').str[0]
# using lambda functions
final_df['Arrival_Time']=final_df['Arrival_Time'].apply(lambda x:x.split(' ')[0])
```

```python
final_df.head()
```

### Arrival hour and minute split

```python
#hour split
final_df['Arrival_hour']=final_df['Arrival_Time'].apply(lambda x:x.split(':')[0])
#minute split
final_df['Arrival_minute']=final_df['Arrival_Time'].apply(lambda x:x.split(':')[1])
```

### convert hour and minute to int

```python
final_df['Arrival_hour']=final_df['Arrival_hour'].astype(int)
final_df['Arrival_minute']=final_df['Arrival_minute'].astype(int)
final_df.info()
```

```python
final_df.drop('Arrival_Time',axis=1,inplace=True)
final_df.head()
```

### Departure time split into minute and hour and convert to int

```python
# hour split
final_df['Dept_hour']=final_df['Dep_Time'].apply(lambda x:x.split(':')[0])
# minute split
final_df['Dept_minute']=final_df['Dep_Time'].apply(lambda x:x.split(':')[1])

#int conversion
final_df['Dept_hour']=final_df['Dept_hour'].astype(int)
final_df['Dept_minute']=final_df['Dept_minute'].astype(int)

#delete dept_time from df
final_df.drop('Dep_Time',axis=1,inplace=True)

final_df.head()
```

```python
final_df.info()
```

```python
final_df['Total_Stops'].unique()
```

```python
final_df['Total_Stops']=final_df['Total_Stops'].map({'non-stop':0,'1 stop':1,'2 stops':2,'3 stops':3,'4 stops':4,'nan':1})
```

```python
final_df.drop('Route',axis=1,inplace=True)
final_df.head()
```

```python
final_df['Additional_Info'].unique()
```

```python
final_df.info()
```

```python
final_df['duration_hour']=final_df['Duration'].apply(lambda x:x.split('h')[0])
final_df.head()
```

```python
final_df.info()
```

```python
#drop rows where hour consist of 5m
final_df.drop([6474,2660],inplace=True)
```

```python
# final_df[final_df['duration_hour']=='5m']
final_df['duration_hour']=final_df['duration_hour'].astype(int)
```

```python
final_df['duration_hour']*60
```

```python
# for minutes
final_df['duration_minutes']=final_df['Duration'].apply(lambda x:x.split('h')[1]).apply(lambda x:x.split('m')[0])
# final_df.rename(columns={'duration_minute':'duration_minutes'},inplace=True)
```

```python
final_df.drop('Duration',axis=1,inplace=True)
final_df.head()
```

```python
# type(final_df['duration_minutes'])
final_df.info()
```

```python
final_df['duration_minutes']=final_df['duration_minutes'].replace('',0)
final_df['duration_minutes'].unique()
# final_df.head()
```

```python
final_df['duration_minutes']=final_df['duration_minutes'].astype(int)
```
