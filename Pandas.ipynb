```python
# load iris dataset from seaborn library
import seaborn as sns
df=sns.load_dataset('iris')
```

```python
#first five rows
df.head()
```

```python
# last five rows
df.tail()
```

```python
#count of rows and columns
df.shape
```

```python
# aggregate values of stats
df.describe()
```

```python
# individual rows/columns
# row extraction= iloc[]
df.iloc[:3,:2]
```

```python
#column extraction = Loc()
df.loc[:3,('sepal_length','sepal_width')]
```

```python
# to remove a column, axis=1
df.drop('species',axis=1,inplace=True)
```

```python
# to remove a row,  axis=0
df.drop([1,2,3],axis=0)
```

```python
# to find the mean
df.mean()
```
